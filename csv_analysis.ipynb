```python
import pandas as pd

data1= pd.read_csv('C:/rsrch/240315_ijepa/logs/vith14.224-bs.2048-ep.300/jepa_r0.csv')
data1=data1.loc[2:,]

# Convert string elements to float in data1 dataframe
data1 = data1.apply(pd.to_numeric, errors='coerce')

# Calculate moving average of loss
data1['moving_avg'] = data1['loss'].rolling(window=10).mean()

'''
Input to Copilot:
In data1 for each epoch, draw line plot of which the x-axis is 'itr' column and the y-axis is 'moving_avg' column. Show all line plots in the same plane.
'''

import matplotlib.pyplot as plt

# Group data by epoch
grouped_data = data1.groupby('epoch')

# Create a figure and axis
fig, ax = plt.subplots()

# Iterate over each group and plot line plot
for epoch, group in grouped_data:
    ax.plot(group['itr'], group['moving_avg'], label=f'Epoch {epoch}')

# Set labels and title
ax.set_xlabel('itr')
ax.set_ylabel('moving_avg')
ax.set_title('Line Plots for Each Epoch')

# Show the legend
ax.legend()

# Show the plot
plt.show()
```
